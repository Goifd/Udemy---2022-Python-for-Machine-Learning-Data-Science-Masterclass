## Merge = Join(SQL)
how to deal with information only present in one of the joined tables:
* inner - present in both tables
* outer
* left or right

```python
import pandas as pd
import numpy as np
registrations = pd.DataFrame({'reg_id':[1,2,3,4],'name':['Andrew','Bobo','Claire','David']})
logins = pd.DataFrame({'log_id':[1,2,3,4],'name':['Xavier','Andrew','Yolanda','Bobo']})
```

```python
registrations
```

```python
logins
```

```python
 # help(pd.merge)
inner_merge=pd.merge(registrations,logins,how='inner',on='name')
```

```python
inner_merge.iloc[0]
```

```python
pd.merge(left=registrations,right=logins,how='left',on='name')
```

```python
pd.merge(left=registrations,right=logins,how='right',on='name') # roder of inputs does matter
```

```python
pd.merge(registrations,logins,how='outer',on='name')
```

```python
registrations=registrations.set_index('name')
```

```python
registrations
```

```python
pd.merge(registrations,logins,left_index=True,right_on='name',how='inner')
```

```python
registrations=registrations.reset_index()
```

```python
registrations
```

```python
registrations.columns=['reg_name','reg_id']
```

```python
registrations
```

```python
pd.merge(registrations,logins,how='inner',left_on='reg_name',right_on='name').drop('reg_name',axis=1)
```

```python
registrations.columns=['name','id']
```

```python
logins.columns=['id','name']
```

```python
pd.merge(registrations,logins,how='inner',on='name',suffixes=('_reg','_log'))
```
