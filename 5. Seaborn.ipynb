# Distribution plots
* rug plot (y axis not interpretable
* histogram -> KDE function (estimate continuous probabiliry curve on data)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\hehexd\\Desktop\\Drive\\Untitled Folder\\UNZIP_FOR_NOTEBOOKS_FINAL\\05-Seaborn\\dm_office_sales.csv")
```

```python
df.head()
```

```python
sns.rugplot(x=df["salary"],data=df,height=0.3)
```

```python
# high level, sets the whole notebook sns.set(style='darkgrid')
sns.displot(data=df,x='salary',bins=40,color='red',edgecolor='blue',kde=True,rug=True)
```

```python
sns.histplot(data=df,x='salary')
```

```python
sns.kdeplot(data=df,x='salary')
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0,100,200)
sample_ages = pd.DataFrame(sample_ages,columns=['age'])
```

```python
sns.rugplot(data=sample_ages,x='age',height=0.5)
```

```python
sns.displot(data=sample_ages,x='age',rug=True,bins=50,kde=True)
```

```python
sns.kdeplot(data=sample_ages,x='age',clip=[0,100],bw_adjust=1,shade=True)
```

# Categorical Plots

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

# Distribution with categories
* Boxplot

```python
df = pd.read_csv("C:\\Users\\hehexd\\Desktop\\Drive\\Untitled Folder\\UNZIP_FOR_NOTEBOOKS_FINAL\\05-Seaborn\\StudentsPerformance.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df,y='math score',x='parental level of education',hue='test preparation course')
plt.legend(bbox_to_anchor=(1.05,1.05))
```

```python
plt.figure(figsize=(12,10))
sns.violinplot(data=df,x="reading score",y='parental level of education',hue="test preparation course",split=True,inner='quartile')
```

```python
sns.boxenplot(x='math score',y='test preparation course',data=df)
```

# Comparison plots
* jointplot()
    * scatter plot + histogram
* pairplot()
    * quick way to compare all the numerical columns in a data frame
    * can be cpu and ram intensive

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\hehexd\\Desktop\\Drive\\Untitled Folder\\UNZIP_FOR_NOTEBOOKS_FINAL\\05-Seaborn\\StudentsPerformance.csv")
```

```python
df.head()
```

```python
sns.jointplot(data=df,x='math score',y='reading score',hue="gender") # kind is default scatter
```

```python
sns.pairplot(df,hue='gender',corner=True) # grabs all the numerical columns automatically
```

# Seaborn grids (matplotlib subplot)
* catplot

```python
sns.catplot(data=df,x='gender',y='math score',kind='box',
           row='lunch',col='test preparation course')
```

```python
g = sns.PairGrid(df,hue='gender')
g = g.map_upper(sns.scatterplot)
g = g.map_lower(sns.kdeplot)
g = g.map_diag(sns.histplot)
g.add_legend()
```
