```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# Functional method
x = np.arange(0,10,0.5)
y = x**2
plt.plot(x,y) # this also returns an object
plt.xlabel('x')
plt.ylabel('y')
plt.title("Title")
plt.xlim(0,10)
plt.ylim(0,100)
plt.show()
plt.savefig('myfirstplot.png')
```

```python
# OOP method
# create figure object, blank canvas
fig = plt.figure(figsize=(10,10))
# Add axis to the figure with 
axes = fig.add_axes([0,0,0.5,0.5])
# plot on the axis
axes.plot(x,y)
```

```python
a=np.linspace(0,10,11)
b=a**4
```

```python
x=np.arange(0,10,1)
y = 2*x
```

```python
fig = plt.figure()
# LARGE AXES
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)
axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('a')
axes1.set_ylabel('b')
axes1.set_title("Power of 4")
# SMALL AXES
axes2= fig.add_axes([0.2,0.5,0.25,0.25]) # x,y of bottom left, width, height
axes2.plot(x,y)
axes2.set_xlim(0,2)
axes2.set_ylim(0,50)
axes2.set_xlabel('a')
axes2.set_ylabel('b')
axes2.set_title("Power of 4")

plt.show()
```

```python
fig = plt.figure(figsize=(2,2),dpi=200)
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

fig.savefig("new_figure.png",bbox='tight')
```

```python
# axes is a numpy array holding the individual axis objects
# fig is the figure canvas
# two function returns a tuple with a fig object and a list of axis objects
fig, axes = plt.subplots(nrows=3,ncols=1,figsize=(4,10))
axes[0].plot(x,y)
axes[0].set_xlabel('x')
axes[1].plot(a,b)
# for no overlaps plt.tight_layout()
'''
for ax in axes:
    ax.plot(x,y)
'''
fig.subplots_adjust(wspace=0.9,hspace=0.9)

fig.suptitle("super title",fontsize=15)
fig.set_figwidth(10)

fig.savefig('newplot.png',bbox_inches='tight')
```
